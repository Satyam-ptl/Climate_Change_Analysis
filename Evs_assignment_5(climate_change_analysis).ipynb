```python
import pandas as pd
df = pd.read_csv("abc.csv")
df.drop(df.columns[1],axis=1,inplace = True)
df.head()
```

```python
# Melt the wide-format year columns into long format
df_melted = df.melt(id_vars='States', var_name='Year', value_name='Temperature')

# Convert year to int
df_melted['Year'] = df_melted['Year'].astype(int)

df_melted.head()
```

```python
import matplotlib.pyplot as plt
# Group by year to get India's average temperature
df_national = df_melted.groupby('Year')['Temperature'].mean()

plt.figure(figsize=(12, 6))
plt.plot(df_national.index, df_national.values, color='brown', linewidth=3)
plt.title('India Average Temperature Trend (1901–2020)',color = "red",size = 30)
plt.xlabel('Year',color = "blue",size = 25)
plt.ylabel('Temperature (°C)',color="blue",size = 25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
states_to_plot = ['Maharashtra', 'Kerala', 'Delhi']  # change as needed

plt.figure(figsize=(12, 6))
for state in states_to_plot:
    state_data = df_melted[df_melted['States'] == state]
    plt.plot(state_data['Year'], state_data['Temperature'], label=state)

plt.title('Temperature Trend by State (1901–2020)',size = 30)
plt.xlabel('Year',color = "brown",size = 25)
plt.ylabel('Temperature (°C)',color = "brown",size = 25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
rolling_avg = df_national.rolling(window=10).mean()

plt.figure(figsize=(12, 6))
plt.plot(df_national.index, df_national.values, label='Annual Avg', alpha=0.4)
plt.plot(rolling_avg.index, rolling_avg.values, label='10-Year Rolling Avg', color='red', linewidth=2)
plt.title('Smoothed National Temperature Trend (1901–2020)',color = "brown",size = 30)
plt.xlabel('Year',color = "brown",size =25)
plt.ylabel('Temperature (°C)',color = "brown",size = 25)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
